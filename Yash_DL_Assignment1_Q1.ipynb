```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
```

```python
'''fashion_mnist.load_data(): This function is likely from a machine learning library like TensorFlow or Keras,
and it's used to load the Fashion MNIST dataset. Fashion MNIST is a dataset of Zalando's article images—consisting
of 60,000 training examples and 10,000 test examples. Each example is a grayscale image of a clothing item from one of 10 categories.'''


# Load the Fashion MNIST dataset into x_train and y_train
(x_train, y_train), (_, _) = fashion_mnist.load_data()

# Define the names of the classes
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# unique_values = set(y_train)
# print("Different values in y_train:", unique_values)
# print(class_names[1])


# Create a dictionary to store one example for each class
examples_by_class = {}

# Find one example for each class
for class_index in range(len(class_names)):
    class_indices = np.where(y_train == class_index)[0]
    example_index = class_indices[0]
    examples_by_class[class_index] = x_train[example_index]


# Plot one example for each class
plt.figure(figsize=(10, 5))
for class_index, example in examples_by_class.items():
    plt.subplot(2, 5, class_index + 1)
    '''In matplotlib, subplots are indexed starting from 1, not 0.
    Therefore, when iterating over the class_index, which starts from 0,
    we need to add 1 to it to get the correct subplot index within the grid.'''
    plt.imshow(example, cmap='gray')
    plt.title(class_names[class_index])
    plt.axis('off')
plt.show()
```
